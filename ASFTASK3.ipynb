```
import matplotlib.pyplot as pl
import pandas as pd
df = pd.read_csv("student.csv")
Name=df['Student Name']
Study_Hours=df['Study hours']

pl.pie(Study_Hours,labels=Name,shadow=True,autopct="%1.1f%%")

pl.title("Students with their Study Hours")
pl.legend()
```
